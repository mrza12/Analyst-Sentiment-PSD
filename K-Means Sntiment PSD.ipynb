## **CLEANING**

```python
import re
import string
import nltk

# Fungsi untuk menghapus URL
def remove_URL(tweet):
    url = re.compile(r'https?://\S+|www\.\S+')
    return url.sub(r'', tweet)

# Fungsi untuk menghapus tag HTML
def remove_html(tweet):
    html = re.compile(r'<.*?>')
    return html.sub(r'', tweet)

# Fungsi untuk menghapus emoji
def remove_emoji(tweet):
    emoji_pattern = re.compile("["
                              u"\U0001F600-\U0001F64F"  # emoticons
                              u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                              u"\U0001F680-\U0001F6FF"  # transport & map symbols
                              u"\U0001F1E0-\U0001F1FF"
                              "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', tweet)

# Fungsi untuk menghapus angka
def remove_numbers(tweet):
    tweet = re.sub(r'\d+', '', tweet)
    return tweet

# Fungsi untuk menghapus simbol dan nama pengguna, dan mengubah kata-kata tertentu
def remove_symbols_and_usernames(tweet):
    tweet = re.sub(r'@\w+', '', tweet)  # Menghapus nama pengguna
    tweet = re.sub(r'[^a-zA-Z0-9\s]', '', tweet)  # Menghapus semua simbol

    # Mengubah kata-kata tertentu
    tweet = re.sub(r'\byg\b', 'yang', tweet)
    tweet = re.sub(r'\btak\b', 'tidak', tweet)
    tweet = re.sub(r'\bgak\b', 'tidak', tweet)
    tweet = re.sub(r'\bnggak\b', 'tidak', tweet)
    tweet = re.sub(r'\bgimana\b', 'bagaimana', tweet)
    tweet = re.sub(r'\bgmn\b', 'bagaimana', tweet)
    tweet = re.sub(r'\bbs\b', 'bisa', tweet)
    tweet = re.sub(r'\btp\b', 'tetapi', tweet)
    tweet = re.sub(r'\bdri\b', 'dari', tweet)
    tweet = re.sub(r'\bdlm\b', 'dalam', tweet)
    tweet = re.sub(r'\bu\b', 'kamu', tweet)
    tweet = re.sub(r'\bOrg\b', 'orang', tweet)
    tweet = re.sub(r'\bsbb\b', 'sebagainya', tweet)
    tweet = re.sub(r'\bngga\b', 'tidak', tweet)

    return tweet

# Membersihkan teks menggunakan fungsi-fungsi yang telah dibuat
df['cleansing'] = df['full_text'].apply(lambda x: remove_URL(x))
df['cleansing'] = df['cleansing'].apply(lambda x: remove_html(x))
df['cleansing'] = df['cleansing'].apply(lambda x: remove_emoji(x))
df['cleansing'] = df['cleansing'].apply(lambda x: remove_numbers(x))
df['cleansing'] = df['cleansing'].apply(lambda x: remove_symbols_and_usernames(x))

df.head(50)
```

## **CASE FOLDING**

```python
def case_folding(text):
  if isinstance(text, str):
    lowercase_text = text.lower()
    return lowercase_text
  else:
    return text

df['case_folding'] = df['cleansing'].apply(case_folding)

df.head(50)
```

## **TOKENIZATION**

```python
def tokenize(text):
  tokens = text.split()
  return tokens

df['tokenize'] = df['case_folding'].apply(tokenize)

df.head()
```

# **filterring/stopword removal**

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words('indonesian')
```

```python
def remove_stopwords(text):
  return[word for word in text if word not in stop_words]

df['filtering/stopword removal'] = df['tokenize'].apply(lambda x: remove_stopwords(x))
df
```

# **K-Means Clustering Prosses**

```python
import nltk
nltk.download('stopwords')

import re
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from textblob import TextBlob  # Add this import statement
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD

# Contoh penggunaan TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['stemming_data'])

# Reduksi dimensi menggunakan Truncated SVD (opsional)
svd = TruncatedSVD(n_components=100)
tfidf_matrix = svd.fit_transform(tfidf_matrix)
```

### **Menentukan jumlah cluster menggunakan elbow**

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Pencarian jumlah kluster yang optimal (contoh dengan metode elbow)
inertia_values = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(tfidf_matrix)
    inertia_values.append(kmeans.inertia_)

# Visualisasi metode elbow
plt.plot(k_values, inertia_values, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Menerapkan k-means dengan jumlah kluster yang telah dipilih
chosen_k = 3  # Gantilah dengan jumlah kluster yang Anda pilih dari metode elbow
kmeans = KMeans(n_clusters=chosen_k, random_state=42)
df['cluster_label'] = kmeans.fit_predict(tfidf_matrix)
df
```

```python
# 1. Eksplorasi dan Interpretasi Hasil Klustering
for cluster_label in range(3):
    print(f"\nCluster {cluster_label}:")
    cluster_data = df[df['cluster_label'] == cluster_label]['stemming_data']
    print(cluster_data.head())
```

```python
# 2. Visualisasi Hasil Klustering (Word Cloud)
from wordcloud import WordCloud

for cluster_label in range(chosen_k):
    cluster_text = ' '.join(df[df['cluster_label'] == cluster_label]['stemming_data'])
    wordcloud = WordCloud(width=800, height=400, max_words=200, background_color='white').generate(cluster_text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'Word Cloud for Cluster {cluster_label}')
    plt.axis('off')
    plt.show()
```

```python
def getPolarity(text):
    analysis = TextBlob(text)
    try:
        analysis = analysis.translate(to="en")
    except Exception as e:
        print(e)

    return analysis.sentiment.polarity

df['polarity'] = df['stemming_data'].apply(getPolarity)
df
```

```python
def getSubjectivity(text):
    analysis = TextBlob(text)
    try:
        analysis = analysis.translate(to="en")
    except Exception as e:
        print(e)

    return analysis.sentiment.subjectivity

df['subjectivity'] = df['stemming_data'].apply(getSubjectivity)
df
```

```python
def getClassification(val):
    if val < 0:
        return 'negatif'
    elif val == 0:
        return 'netral'
    else:
        return 'positif'
df['sentimen'] = df['polarity'].apply(getClassification)
df
```
